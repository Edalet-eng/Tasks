### Pandas kitabxanası ilə sütun adlarının dəyişdirilməsi

```python
# Lazım olan kitabxanalar
import seaborn as sns
import pandas as pd
```

```python
# Titanik datasetinin yüklənməsi
df = sns.load_dataset(name = 'titanic')

# İlk beş sətirin göstərilməsi
df.head()
```

```python
# Sütun adlarının ilk hərfinin capitalize() metodu ilə böyüdülməsi
df.columns = df.columns.str.capitalize()

# İlk beş sətirin göstərilməsi
df.head()
```

```python
# Sütun adlarının lower() metodu ilə tamamilə kiçik hərflərlə əvəzlənməsi
df.columns = df.columns.str.lower()

# İlk beş sətirin göstərilməsi
df.head()
```

```python
# sex sütununun replace() metodundan istifadə edərək gender sütunu adı ilə əvəz olunması
df.columns = df.columns.str.replace(pat = 'sex', repl = 'gender')

# İlk beş sətirin göstərilməsi
df.head()
```

```python
# ticket_price və gender sütunlarının replace() metodundan istifadə edərək əvvəlki sütunu adları ilə əvəz olunması
df.rename(columns = {'gender':'sex', 'fare':'ticket_price'}, inplace = True)

# İlk beş sətirin göstərilməsi
df.head()
```

```python
# Sütun adlarında potensial boşluqların strip() metodu ilə silinməsi
df.columns = df.columns.str.strip()
# İlk beş sətirin göstərilməsi
df.head()
```

```python
# Bulean data növlü sütunların dəyərlərinin applymap() metodundan istifadə edərək int() metodu və adsız funksiya ilə integer dəyərlərlə (True --> 1 & False --> 0) əvəzlənməsi
df[['adult_male', 'alone']] = df[['adult_male', 'alone']].applymap(func = lambda x: int(x))

# İlk beş sətirin göstərilməsi
df.head()
```

```python
# Strinqlərdən ibarət list data strukturunun dates dəyişəninə qeyd olunması
dates = ['2021-12-31', '2022-01-01', '2022-02-01', '2022-03-31', '2022-04-01', '2022-05-01']

# dates dəyişəninin dəyərlərinin zaman dəyərlərinə çevrilməsi
dates = pd.to_datetime(arg = dates, yearfirst = True)

# dates dəyişəninin dəyər və tarix strinqinin isə açar oldugu lügət data strukturunun yaradılması
data_dictionary = {'tarix':dates}

# data_dictionary dəyişəninin pd.DataFrame() metodu vasitəsilə dataframe data strukturunun yaradılması
df = pd.DataFrame(data = data_dictionary)
df
```

```python
# year atributu ilə tarix sütunundan illərin çıxardılması
df['il'] = df.tarix.dt.year
df
```

```python
# month atributu ilə tarix sütunundan ayların çıxardılması
df['ay'] = df.tarix.dt.month
df
```

```python
# month_name() metodu ilə tarix sütunundan ay adlarının çıxardılması
df['ay_adı'] = df.tarix.dt.month_name()
df
```

```python
# day atributu ilə tarix sütunundan günlərin çıxardılması
df['gün'] = df.tarix.dt.day
df
```

```python
# day_name() metodu ilə tarix sütunundan gün adlarının çıxardılması
df['gün_adı'] = df.tarix.dt.day_name()
df
```

```python
# quarter atributu ilə tarix sütununda olan ayın həmin ilin hansı rübünə düşdüyünün göstərilməsi
df['rüb'] = df.tarix.dt.quarter
df
```

```python
# pd.DateOffset() metodu və years parametri vasitəsi ilə bir il öncəki tarixin tapılması
df['1_il_öncəki_tarix'] = df.tarix - pd.DateOffset(years = 1)
df
```

```python
# pd.DateOffset() metodu və months parametri vasitəsi ilə altı ay öncəki tarixin tapılması
df['6_ay_öncəki_tarix'] = df.tarix- pd.DateOffset(months = 6)
df
```

```python
# pd.DateOffset() metodu və days parametri vasitəsi ilə 4 gün öncəki tarixin tapılması
df['4_gün_öncəki_tarix'] = df.tarix - pd.DateOffset(days = 4)
df
```

```python
# pd.DateOffset() metodu və weeks parametri vasitəsi ilə 2 həftə öncəki tarixin tapılması
df['2_həftə_sonraki_tarix'] = df.tarix + pd.DateOffset(weeks = 2)
df
```

```python
# drop_duplicates() metodundan istifadə edərək dublikat sətirlərin silinməsi
df.drop_duplicates(inplace = True, ignore_index = True)
print(f'Dublikat sətirlərin sayı: {df.duplicated().sum()}')
```
