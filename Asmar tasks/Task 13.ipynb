
# Qruplaşdırma və Qrup Statistikaları:

```python
# DataFrame-də müəyyən bir sütuna əsasən məlumatları qruplaşdırın və hər qrupun orta qiymetini hesablayın.
df.groupby(by='Şəhər')['Qiymet'].mean()
```

```python
# modeline ve seherine gore qiymetin ve baxis sayinin min/max qiymetlerini versin
df.groupby(by =['Model','Şəhər'])[['Qiymet', 'Baxışların sayı']].agg(func=['max', 'min'])
```

```python
# eyniylə yuxaridaki kimi quruplasdir amma qiymetden max, baxis-dan min tap
df.groupby(by=['Model','Şəhər'])[['Qiymet','Baxışların sayı']].agg(func={'Qiymet':'max','Baxışların sayı':'min'})
```
