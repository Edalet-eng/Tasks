
Generate a list of 100 random numbers in range from 0 to 9 and display the amount of occurances of each point

```python
numbers=np.random.randint(0, 10, 100)
tekrarlanma= np.bincount(numbers)

plt.bar(range(10), tekrarlanma)
plt.xlabel('Ededler')
plt.ylabel('Tekrarlanma miqdari')
plt.title('Qrafik')
plt.show()
```


Plot 4 points in the shape of a square

```python
x = [1, 1, 4 , 4]
y = [1, 4, 4, 1]

plt.plot(x, y,  color='b', markerfacecolor='red', markeredgecolor='black')
plt.plot(x+[x[0]],y+[y[0]])
plt.title('Kvadrat')
plt.show()
```

Plot a circle

```python
bucaq= np.linspace(0, 2 * np.pi, 100)
r = 1
x = r * np.cos(bucaq)
y = r * np.sin(bucaq)
plt.plot(x, y)
plt.xlabel('x oxu')
plt.ylabel('y oxu')
plt.show()
```

Plot a quadratic function

```python
x = np.linspace(0, 10, 20)
y = x**2 - 6*x + 5
plt.plot(x, y)
plt.title('Kvadratik')
plt.show()
```

Plot a cubic function

```python
x = np.linspace(0, 10, 20)
y = 2*x**3 - 3*x**2 + x - 5
plt.plot(x, y)
plt.title('Kubik')
plt.show()
```

Plot exponential function

```python
x = np.linspace(0, 10, 20)
y = np.exp(x)
plt.plot(x, y)
plt.title('Eksponential')
plt.show()
```

Plot sigmoid function

```python
x = np.linspace(0, 10, 20)
y = 1/(1 + np.exp(-x))
plt.plot(x, y)
plt.title('sigmoid')
plt.show()
```

Plot simusoidal function

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

Plot cotangent function

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.cos(x) / np.sin(x)
plt.plot(x, y)
plt.show()
```
