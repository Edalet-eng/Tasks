Create a recursive function that finds a sum of all elements in a nested list

Example

[1, 2, [3,4], [5,6]] -> 1 + 2 + 3 + 4 + 5 + 6 = 21

```python
def func(a):
    cem=0
    for i in a:
        if type(i)==int:
            cem+=i
        else:
            for x in i:
                cem+=x
    return cem
    
    
a=[1, 2, [3,4], [5,6]]
func(a)
```
