```python
df = df.fillna({'INT_SQFT': df['INT_SQFT'].median()})
df
```

```python
df = df.fillna({'DIST_MAINROAD': df['DIST_MAINROAD'].mean()})
df
```

```python
a= df['N_BATHROOM'].median()
df['N_BATHROOM'].fillna(a)
```

```python
df = df.fillna({'N_BATHROOM': df['N_BATHROOM'].mode()})
df
```

```python
df = df.fillna({'N_ROOM': df['N_ROOM'].median()})
df
```

```python
df['DATE_SALE'] = pd.to_datetime(df['DATE_SALE'])
df['Day'] = df['DATE_SALE'].dt.day
df['Month'] = df['DATE_SALE'].dt.month
df['Year'] = df['DATE_SALE'].dt.year

# DATE_SALE sütununu düşürmək
df.drop(columns=['DATE_SALE'], inplace=True)
df
```
