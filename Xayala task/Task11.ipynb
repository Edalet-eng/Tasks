Create an arbitrary dataframe with two columns a, b that contain numbers. Create the c column which equals to the multiplication of values of column a to column b.

```python
import pandas as pd

data = {
    'a': [1, 2, 3, 4, 5],
    'b': [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data)

df['c'] = df['a'] * df['b']

df
```

Find out the alcohol consumption details in the year '1986' or '1989' where WHO region is 'Americas' or 'Europe' from the world alcohol consumption dataset

```python
grouped = df.groupby(by=['WHO region', 'Year'])['Display Value'].sum().reset_index()

region = (grouped['WHO region'] == 'Americas') | (grouped['WHO region'] == 'Europe')

year = (grouped['Year'] == 1986) | (grouped['Year'] == 1989)

region_year= grouped[region & year]

region_year
```
