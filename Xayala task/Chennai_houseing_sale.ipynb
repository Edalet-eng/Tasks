```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Chennai_houseing_sale.csv")
df
```

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
df
```

```python
df.info()
```

```python
df['date_sale'] = df['date_sale'].str.replace('-', '/')
df['date_sale'].str.split('/')
df['date_sale_day']=df['date_sale'].apply(lambda x: x.split('/') [0])
df['date_sale_month']=df['date_sale'].apply(lambda x: x.split('/') [1])
df['date_sale_year']=df['date_sale'].apply(lambda x: x.split('/') [2])
df = df.drop(columns=['date_sale'])
df.head(6)
```

```python
df['date_build'] = df['date_build'].str.replace('-', '/')
df['date_build'].str.split('/')
df['date_build_day']=df['date_build'].apply(lambda x: x.split('/') [0])
df['date_build_month']=df['date_build'].apply(lambda x: x.split('/') [1])
df['date_build_year']=df['date_build'].apply(lambda x: x.split('/') [2])
df = df.drop(columns=['date_build'])
df.head(6)
```

```python
df.isna().sum()
```

```python
 df['n_bedroom'].value_counts()
```

```python
median_value = df['n_bedroom'].median()
median_value
```

```python
mean_value = df['n_bedroom'].mean()
mean_value
```

```python
mode_value = df['n_bedroom'].mode()[0]
df['n_bedroom'] = df['n_bedroom'].fillna(mode_value)
```

```python
 df['n_bathroom'].value_counts()
```

```python
mode_value = df['n_bathroom'].mode()[0]
df['n_bathroom'] = df['n_bathroom'].fillna(mode_value)
```

```python
 df['qs_overall'].value_counts()
```

```python
mean_value = df['qs_overall'].mean()
df['qs_overall'] = df['qs_overall'].fillna(mean_value)
```

```python
df.info()
```

```python
df['n_bedroom'] = df['n_bedroom'].astype(int)
df['n_bathroom']=df['n_bathroom'].astype(int)
df['park_facil'] = df['park_facil'].astype(bool)
df.dtypes
```

```python
df = df.drop(columns=['prt_id'])
df = df.drop(columns=['area'])
df = df.drop(columns=['sale_cond'])
df = df.drop(columns=['buildtype'])
df = df.drop(columns=['date_sale_day'])
df = df.drop(columns=['date_sale_month'])
df = df.drop(columns=['date_sale_year'])
df = df.drop(columns=['date_build_day'])
df = df.drop(columns=['date_build_month'])
df = df.drop(columns=['date_build_year'])
df = df.drop(columns=['park_facil'])
df = df.drop(columns=['utility_avail'])
df = df.drop(columns=['street'])
df = df.drop(columns=['mzzone'])
df
```

```python
numeric_columns = ['int_sqft', 'dist_mainroad', 'n_bedroom', 'n_bathroom', 'n_room','qs_rooms','qs_bathroom','qs_bedroom','qs_overall',
                   'reg_fee', 'commis']
for column in numeric_columns:
    plt.boxplot(df[column])
    plt.title(f'Box Plot of {column}')
    plt.show()
```

```python
for column in numeric_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    outliers = df[(df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))]
    print(f'Outliers in {column}:')
    print(outliers[column])
```

```python
for column in numeric_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    df = df[~((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR)))]
df
```
