Satici,Telefonlar,Ötürücü,Baxışların_sayı,url,Sahiblər,Extra,Etrafli,Yeni,Yeniləndi,Hansı bazar üçün,Yerlərin_sayı sütunlarını datasetdən sil.

```python
columns_to_drop = [
    'satici', 'telefonlar', 'baxışların_sayı', 'url',
    'sahiblər', 'extra', 'etrafli', 'yeni', 'yeniləndi',
    'hansı_bazar_üçün_yığılıb', 'yerlərin_sayı'
]

df.drop(columns=columns_to_drop, inplace=True)
df.head(5)
```

Mühərrik sütununu parçala(mühərrik həcmi, mühərrik gücü, yanacaq növü) sütunlarına ayır
.

```python
df['mühərrik'].str.split('/')
df['mühərrik hecmi']=df['mühərrik'].apply(lambda x: x.split('/') [0])
df['mühərrik gücü']=df['mühərrik'].apply(lambda x: x.split('/') [1])
df['yanacaq növü']=df['mühərrik'].apply(lambda x: x.split('/') [2])
df = df.drop(columns=['mühərrik'])
df.head(6)
```

mühərrik sütunununda L-i sil
at_gücü sütunununda a.g-i sil
yürüş sütunununda km-i sil

```python
df['mühərrik hecmi'] = df['mühərrik hecmi'].str.replace('L', '')

df['mühərrik gücü'] = df['mühərrik gücü'].str.replace('a.g.', '')
df['yürüş'] = df['yürüş'].str.replace('km', '')
df.head(5)
```

mühərrik sütununun data növünü float data növünə dəyiş
at_gücü və yürüş sütunlarının data növünü int data növünə dəyiş

```python
df['mühərrik gücü'] = df['mühərrik gücü'].astype(int)
df['mühərrik hecmi']=df['mühərrik hecmi'].astype(float)
df['yürüş'] = df['yürüş'].str.replace(' ','')
df['yürüş'] = df['yürüş'].astype(int)
df.dtypes
```

Cari ildən buraxılış_ili sütununun dəyərlərini çixib yaş sütunu yarat

```python
cari_il = 2024

df['yaş'] = cari_il - df['buraxılış_ili']

df.yaş
```

qiymət_kateqoriyası və qiymət_faizi sütunlarını yarat. qiymət_kateqoriyası sütunu üçün bu intervaldan və kateqoriyalardan istifadə et.
Bins = [0, 10000, 30000, 50000, 150000, 300000, qiymət sütununun maksimum dəyəri]

Labels = ['Çox Ucuz', 'Ucuz', 'Normal', 'Bahalı', 'Çox Bahalı', 'Lyüks']

```python
max_price = df['qiymet'].max()
bins = [0, 10000, 30000, 50000, 150000, 300000, max_price]
labels = ['Çox Ucuz', 'Ucuz', 'Normal', 'Bahalı', 'Çox Bahalı', 'Lyüks']

df['qiymət_kateqoriyası'] = pd.cut(df['qiymet'], bins=bins, labels=labels, right=False)

df['qiymət_faizi'] = (df['qiymet'] / max_price) * 100

df[['qiymet', 'qiymət_kateqoriyası', 'qiymət_faizi']]
```
