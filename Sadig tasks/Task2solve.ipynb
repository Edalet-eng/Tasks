

Create a program that display the surface area and the volume of a sphere. The radius of a sphere is determined by the user input.

Assume π = 3.14.

Example

Radius: 0.5

Surface area: 3.14

Volume: 0.5233333

```python
r = float(input("r: "))
surf_area = 4*3.14*(r**2)
v = 4/3*(3.14*(r**3))
print("Surface area: ",surf_area)
print("Volume: ",v)
```

Create a program that calculates the remaining quantity radioactive substance with the given formula:
    
    
User inputs the initial quantity of the substance, half-life of the substance and time elapsed. Program displays the remaining amount of radioactive substance

Example

Initial quantity of the substance: 10

Half-life of the substance: 5

Time elapsed: 10

The remaining amount of substance: 2.5

```python
n0 = int(input("initial quantity of the substance: "))
t5 = int(input("half-life of the substance: "))
t = int(input("time elapsed: "))
nt = n0*(0.5**(t/t5))
print("The remaining amount of substance: ",nt)
```
